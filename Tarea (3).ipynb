```python
import re 
import csv 
```

```python
with open("pubmed-CRISPRCasn-set.txt", errors="ignore") as l: 
    texto = l.read()
texto = re.sub(r"\n\s{6}", " ", texto)
```

```python
paises_1 = re.findall (r"AD\s{2}-\s[A-Za-z].*,\s([A-Za-z]*)\.\s", texto)
```

```python
paises_1
```

```python
unique_paises = list(set(paises_1))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(countries_lat, countries_long) 
plt.colorbar()
##paises de referencia: USA, UK, Brazil, China, Australia, Norway 
ard = dict(arrowstyle="->")
plt.annotate("USA", xy = (23.424076, 53.847818), 
             xytext = (23.424076, 53.847818), arrowprops = ard) 
plt.annotate("UK", xy= (55.378051, 3.435973), 
            xytext = (55.378051, 3.435973), arrowprops = ard)
plt.annotate("Brazil", xy= (-14.235004, -51.92528), 
            xytext = (-14.235004, -51.92528), arrowprops = ard)
plt.annotate("China", xy= (-35.86166, -104.195397), 
            xytext = (-35.86166, -104.195397), arrowprops = ard)
plt.annotate("Australia", xy= (-25.274398, 133.775136), 
            xytext = (-25.274398, 133.775136), arrowprops = ard)
plt.annotate("Norway", xy= (60.472024, 8.468946), 
            xytext = (60.472024, 8.468946), arrowprops = ard)
params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches( (plSize[0] * 3, plSize[1] * 3) )
plt.show()
```
